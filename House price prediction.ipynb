```python
#Getting the dataset
data=pd.read_csv("housing.csv")
```

```python
data.dropna(inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Splitting the target from features
x=data.drop(['median_house_value'],axis=1)
y=data['median_house_value']
```

```python
y
```

```python
#splitting of training and testing data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
train_data=x_train.join(y_train)
```

```python
train_data
```

```python
#Findind the correlation between the training datasets
train_data.corr()
```

```python
train_data.ocean_proximity.value_counts()
```

```python
train_data['total_rooms']=np.log(train_data['total_rooms']+1)
train_data['total_bedrooms']=np.log(train_data['total_bedrooms']+1)
train_data['population']=np.log(train_data['population']+1)
train_data['households']=np.log(train_data['households']+1)
```

```python
train_data.hist(figsize=(15,8))
```

```python
train_data.ocean_proximity.value_counts()
```

```python
train_data=train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop(['ocean_proximity'],axis=1)
```

```python
train_data
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(),annot=True,cmap='YlGnBu')
```

```python
# Applying the Algorithm in the training data
from sklearn.linear_model import LinearRegression
x_train,y_train=train_data.drop(['median_house_value'],axis=1),train_data['median_house_value']
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
test_data=x_test.join(y_test)
test_data['total_rooms']=np.log(test_data['total_rooms']+1)
test_data['total_bedrooms']=np.log(test_data['total_bedrooms']+1)
test_data['population']=np.log(test_data['population']+1)
test_data['households']=np.log(test_data['households']+1)
test_data=test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(['ocean_proximity'],axis=1)

```

```python
test_data
```

```python
x_test,y_test=test_data.drop(['median_house_value'],axis=1),test_data['median_house_value']
```

```python
# Checking the accuracy of the model
reg.score(x_test,y_test)
```

```python
# Making predictions on the training data and testing data
train_data_prediction=reg.predict(x_train)
```

```python
train_data_prediction
```

```python
test_data_prediction=reg.predict(x_test)
```

```python
test_data_prediction
```

```python
# Comparing the Actual price of the house 
plt.scatter(y_train,train_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices Vs Predicted Prices")
plt.show()
```
